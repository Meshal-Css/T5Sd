### Use appropriate plots such as histograms, bar charts, pie charts, scatter plots, and heatmaps.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Task (2) Dataset.csv')


df['Datetime'] = pd.to_datetime(df['Datetime'])

```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='TrafficIndexLive', kde=True)
plt.title('Distribution of TrafficIndexLive')
plt.xlabel('Traffic Index Live')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
df['Hour'] = df['Datetime'].dt.hour
jams_by_hour = df.groupby('Hour')['JamsCount'].mean().reset_index()
sns.barplot(x='Hour', y='JamsCount', data=jams_by_hour)
plt.title('Average JamsCount by Hour')
plt.xlabel('Hour')
plt.ylabel('Average Jams Count')
plt.show()
```

```python
jams_severity = df['JamsDelay'].apply(lambda x: 'Low' if x < 5 else 'Medium' if x < 15 else 'High')
plt.figure(figsize=(8, 8))
jams_severity.value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Jams Severity Distribution')
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='TrafficIndexLive', y='JamsDelay', data=df)
plt.title('TrafficIndexLive vs. JamsDelay')
plt.xlabel('TrafficIndexLive')
plt.ylabel('JamsDelay (minutes)')
plt.show()
```
